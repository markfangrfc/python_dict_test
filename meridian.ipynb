```python
def Container_of_Tears():
    Chengqi_data = {
        '定位': '在面部，當瞳子外側端直下0.5寸處，鼻翼外緣高位1.5寸處。',
        '解剖': '有眶上神經，眶下神經及淚小管等通過。有眶上動脈的分支分布。鄰近有筋膜及軟組織。',
        '主治': '目眩，目赤腫痛，淚多。',
        '配伍': '配太陽治目赤腫痛;配攢竹治目眩。',
        '刺灸法': '向下平刺0.3～0.5寸，切忌向內刺。'
    }
    return Chengqi_data

def Man_Welcome():
    Renying_data = {
        '定位': '在手臂內側，肘橫紋上3寸處，肱骨內側緣。',
        '解剖': '有肱動脈、肱深動脈及其伴隨靜脈，肱神經與正中神經分布於此。周圍有筋膜及肌肉。',
        '主治': '肩臂疼痛，手臂麻木，肘痛。',
        '配伍': '配側溪治肩痛;配曲池治手臂麻木。',
        '刺灸法': '向上斜刺0.5～1寸，切忌向內刺。'
    }
    return Renying_data

def Empty_Basin():
    Quepen_data = {
        '定位': '在足背側，第一、第二跖骨之間凹陷處。',
        '解剖': '有足背動脈及其伴隨靜脈、淺層脛神經分布。周圍有筋膜及肌肉。',
        '主治': '頭痛，目眩，鼻塞，咽喉腫痛。',
        '配伍': '配攢竹治頭痛;配印堂治目眩。',
        '刺灸法': '向下平刺0.3～0.5寸，切忌向內刺。'
    }
    return Quepen_data

def Heaven_Pivot():
    Tianshu_data = {
        '定位': '在後頸部，頭髮際，後頸外緣，燦笑肌肌中。',
        '解剖': '有較深的頸部筋膜及表浅的皮下組織。周圍有頸動脈的分支及枕動脈分布。',
        '主治': '頭痛，頭暈，目眩，耳鳴，頸項強痛。',
        '配伍': '配風池治頭痛;配大椎治頭暈。',
        '刺灸法': '向下斜刺0.5～1寸，切忌向內刺。'
    }
    return Tianshu_data

def Leg_Three_Miles():
    Zusanli_data = {
        '定位': '在小腿前外側，膝眼下3寸處，脛骨前緣側。',
        '解剖': '有膝前脛深動、膝前脛腓淺動脈及其伴隨靜脈分布。周圍有筋膜及肌肉。',
        '主治': '腹痛，嘔吐，泄瀉，脚氣，膝脛痛。',
        '配伍': '配三陰交治腹痛;配陽陵泉治脚氣。',
        '刺灸法': '向下平刺1～1.5寸，切忌向內刺。'
    }
    return Zusanli_data

def Stomach_Meridian_of_Foot_Yangming():
    acupoints_info = {
        '承泣穴': Container_of_Tears(),
        '人迎穴': Man_Welcome(),
        '缺盆穴': Empty_Basin(),
        '天樞穴': Heaven_Pivot(),
        '足三里穴': Leg_Three_Miles()
        # ... 其他穴位
    }
    return acupoints_info

def get_acupoint_info(acupoint_name, info_category=None):
    acupoints_info = Stomach_Meridian_of_Foot_Yangming()

    acupoint_data = acupoints_info.get(acupoint_name, None)
    if acupoint_data is None:
        return "未找到指定的穴位信息。"

    if info_category:
        return acupoint_data.get(info_category, "未找到指定類別的信息。")

    return acupoint_data

def query_acupoint():
    while True:
        print("\n--- 查詢系統 ---")
        acupoint_name = input("請輸入您要查詢的穴位名稱（輸入'退出'結束查詢）:\n(承泣穴、人迎穴、缺盆穴、天樞穴、足三里穴選擇其一owo):")
        if acupoint_name == "退出":
            print("\n--- 已退出查詢 ---")
            print("感謝使用本系統，祝你好運owo")
            break

        acupoint_info = get_acupoint_info(acupoint_name)
        if "未找到" in acupoint_info:
            print(f"\n{acupoint_info}\n")
            continue

        while True:
            print(f"\n找到穴位：{acupoint_name}")
            print("可查詢的信息類別有：定位、解剖、主治、配伍、刺灸法")
            info_category = input("請輸入您要查詢的信息類別（輸入'全部'顯示所有信息，或輸入'返回'返回上一層）:")

            if info_category == "返回":
                print("\n返回上一層...\n")
                break
            elif info_category == "全部":
                print(f"\n{acupoint_name} - 全部信息:")
                for category, detail in acupoint_info.items():
                    print(f"{category}: {detail}")
                print()  # 為了排版整潔owo
            elif info_category in ["定位", "解剖", "主治", "配伍", "刺灸法"]:
                category_info = get_acupoint_info(acupoint_name, info_category)
                print(f"\n{acupoint_name} - {info_category}:\n{category_info}\n")
            else:
                print("\n輸入的信息類別不正確，請重新輸入。\n")

            # 再次查詢或返回
            while True:
                next_action = input("輸入'再查詢'以查詢此穴位的其他信息，或輸入'返回'返回穴位選擇：")
                if next_action in ["再查詢", "返回"]:
                    break
                else:
                    print("請輸入有效的指令（'再查詢' 或 '返回'）。")
            if next_action == "返回":
                print("\n返回穴位選擇...")
                break
# 啟動查詢
query_acupoint()
```
